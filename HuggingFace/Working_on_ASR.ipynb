```python
from transformers import pipeline
import librosa
import numpy as np

# Initialize the ASR pipeline with the wav2vec2 model
asr_pipeline = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h", device=0)

# Function to load and preprocess the audio file
def load_audio(file_path, target_sr=16000):
    try:
        # Load the audio file with librosa, resample to target sampling rate (16000 Hz)
        y, sr = librosa.load(file_path, sr=target_sr)
        # Normalize the audio to ensure it matches the input format expected by wav2vec2
        y = librosa.util.normalize(y)
        return y, sr
    except Exception as e:
        print(f"Error loading audio file: {e}")
        return None, None

# File path to the audio file
audio_file_path = "/content/Third Audio 2024-10-01 at 00.00.34_f4b31a1e.waptt.opus"

# Load and preprocess the audio file
audio_data, sample_rate = load_audio(audio_file_path)

if audio_data is not None:
    # Perform ASR on the waveform
    audio_input = np.array(audio_data)  # Convert to numpy array if necessary
    output = asr_pipeline(audio_input)

    # Print the transcribed text
    print("Transcribed Text: ", output["text"])
else:
    print("Failed to load audio.")
```
