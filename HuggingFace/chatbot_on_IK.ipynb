```python
# required libraries need to install

import os # for file handling
import nltk # for language process (natural)
from nltk.tokenize import word_tokenize  # convet text into chunks
from nltk.corpus import stopwords # for remove stop words in English
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering, pipeline # call Q&A features, Pipeline for call model, and tokenizer convert text into token/chunks
import torch

# Load the text from IK.txt
path = '/content/IK.txt'

```

```python
# read each line of the files

if os.path.exists(path):
    with open(path) as f:
        lines = f.readlines()
else:
    raise FileNotFoundError(f"File not found: {path}")

```

```python
text = ' '.join(lines) # all lines are combine and convert into string variable text


nltk.download('stopwords')
nltk.download('punkt')

```

```python

text = text.lower() # its convert all text into lower case
text = nltk.RegexpTokenizer(r'\w+').tokenize(text)  # agian text convert into chunks with condition \w+ to pattern identifaction
print(text)
```

```python

# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_text = [word for word in text if word not in stop_words]
```

```python

# Combine filtered tokens back into a string for the model
processed_text = ' '.join(filtered_text)

print(processed_text)
```

```python
# Load pre-trained DistilBERT model for question answering
qa_tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased-distilled-squad')
qa_model = DistilBertForQuestionAnswering.from_pretrained('distilbert-base-uncased-distilled-squad')
```

```python

# Initialize a question-answering pipeline
qa_pipeline = pipeline('question-answering', model=qa_model, tokenizer=qa_tokenizer)
```

```python

# Define a function to get the specific answer to a question
def get_specific_answer(question, context):
    result = qa_pipeline(question=question, context=context)
    return result['answer']


```

```python
# Example usage: Asking a specific question
question = "where Imran Ahmed Khan niazi born and when "
context = ' '.join(lines)  # Use the original text from IK.txt

answer = get_specific_answer(question, context)
print(f"Question: {question}")
print(f"Answer: {answer}")
```
