```python
# first set the model
from transformers import pipeline
pipe = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h",device = 0)
#device = 0 means GPU, device = 1 means GPU 2, DEVICE = -1 MEANS NO GPU ONLY CPU
```

```python
# load the datasets
from datasets import load_dataset
dataset = load_dataset("superb", "asr", split="test")
```

```python
# usek Keydataset libraries
from transformers.pipelines.pt_utils import KeyDataset
```

```python
from tqdm.auto import tqdm
for output in tqdm(pipe(KeyDataset(dataset,'file'))):
    print(output["text"])
```
